There are only 100 different tracks per page so we need to get all the pages.

```python
from random import randint
from time import sleep

def get_playlist_tracks(playlist_id):
    results = sp.user_playlist_tracks("spotify",playlist_id)
    tracks = results['items']
    while results['next']!=None:      # we loop while there's still a new page
        results = sp.next(results)    # we retrieve the result of the next page
        tracks = tracks + results['items'] # we add the new page's tracks to the previous page's track
        sleep(randint(1,3000)/1000)   # we apply the respectful nap
    return tracks
```

```python
# we apply the funtion to our 10.000 songs playlist
all_tracks = get_playlist_tracks("1G8IpkZKobrIlXcVPoSIuf")
```

```python
# we json normalize all the tracks
tracks = json_normalize(all_tracks)
```

Here we can see that we still have a list of several informations for the column 'tracks.artists'.\
We need to add them the different new columns in the dataframe.

```python
# first we convert the lists to dictionnaries since it's easier to jsonnormalize
def list_to_dict(x):
    return {i: x[i] for i in range(len(x))}
            
```

```python
# it seems that there are NaN values in column 'artist_dict', let's check for it
tracks.isna().sum()
```

```python
# tracks['artist_dict'] = tracks['track.artists'].apply(list_to_dict)
# TypeError: object of type 'float' has no len()
```

```python
# we drop the NaN values
tracks=tracks[tracks['track.artists'].isna()==False]
```

```python
tracks.isna().sum()
```

```python
tracks['artist_dict'] = tracks['track.artists'].apply(list_to_dict)
```

```python
tracks
```

```python
tracks['artist_dict'][0]
```

```python
# now we create a dataframe of the column artist_dict we've just created
def expand_list_dict(row):
    df = pd.DataFrame.from_dict(row['artist_dict'], orient='index') # we create the new df from the row in the artist_dict colummn
    df['song_id'] = row['track.id'] # the column wich was 'track.id' in the previous df will be now name 'song_id'
    return df                                              # this will enable us to merge the 2 dfs 

tracks['artists_dfs'] = tracks.apply(expand_list_dict, axis=1)
tracks['artists_dfs'][0]
```

```python
tracks['track.artists'][0]
```

We have the same elements as we had in the column 'track.artists'.

```python
# now we create a new dataframe for all the artist in the column 'artists_dfs'
artist_df = pd.DataFrame(columns=['external_urls', 'href', 'id', 'name', 'type', 'uri']) #we name the column of the new df
for values in tracks['artists_dfs']: # we loop through the column where we want to create the df
    artist_df = pd.concat([artist_df, values], axis=0) # we add the values that we have in the column 'artists_dfs' to the new artist_df we create 
    
artist_df
```

Now we can concat our 2 dataframes so that we have one df with all columns we were looking for.

```python
df_merged = pd.merge(left=tracks,
                    right=artist_df,
                    how='inner',
                    left_on='track.id',
                    right_on='song_id')
df_merged.head()
```

```python
df_merged.shape
```

```python
# now we keep only the columns that we need 
df_final = df_merged[['track.name', 'name', 'song_id']]
df_final
```

```python
# we check for NaN values
df_final.isna().sum()
```

```python
# we delete NaN values
df_final=df_final[df_final['song_id'].isna()==False]
```

```python
df_final.isna().sum()
```

```python
df_final = df_final.reset_index(drop=True)
```

# Get audio features 

```python
%%time
chunks = [(i, i+100) for i in range(0, len(df_final), 100)]
audio_features_list = []
for chunk in chunks:
    id_list100 = df_final['song_id'][chunk[0]:chunk[1]]
    audio_features_list = audio_features_list + sp.audio_features(id_list100)
    sleep(randint(1,3000)/1000)
len(audio_features_list)
```

```python
audio_features_df = json_normalize(audio_features_list)
```

```python
audio_features_df
```

Now we merge this datafram with df_final

```python
# first we need to drop the duplicates
audio_features_df = audio_features_df.drop_duplicates()
```

```python
audio_features_df.shape
```

```python
df_w_audio_ft = pd.merge(left=df_final,
                        right=audio_features_df,
                        how='inner',
                        left_on='song_id',
                        right_on='id')
df_w_audio_ft
```
